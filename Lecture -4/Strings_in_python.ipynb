### Strings are sequence of Characters
In Python specifically, strings are a sequence of Unicode Characters

Topic to be discussed:

1. Creating Strings
2. Accessing Strings
3. Adding Chars to Strings
4. Editing Strings
5. Deleting Strings
6. Operations on Strings
7. String Functions

### Creating Stings

```python
s = 'hello'
s = "hello"
# multiline strings
s = '''hello'''
s = """hello"""
s = str('hello')
print(s)
```

### Accessing Strings

```python
# positive access

s = "welcome to python"
```

```python
print(s[0])
```

```python
print(s[9])
```

```python
# Negative indexing

s = "welcome to python"
```

```python
print(s[-9])
```

```python
print(s[-5])
```

```python
# String slicing

s = "welcome to python"
```

```python
print(s[0:7])
```

```python
print(s[11:17])
```

```python
print(s[11:500])
```

```python
print(s[11:])
```

```python
print(s[:])
```

```python
print(s[6::-1])
```

```python
print(s[::-1])
```
