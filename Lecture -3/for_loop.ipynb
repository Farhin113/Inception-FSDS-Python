#### Exercise ; print triangle coordinates

```python
row =int(input("Enter the number of rows: "))

for i in range(1,row+1):
    for j in range(1,i+1):
        print("*",end="")

    print(" ")    
    
```

```python
row =int(input("Enter the number of rows: "))

for i in range(row,0,-1):
    for j in range(i,0,-1):
        print("*",end="")
       
    print(" ")    
    
```

### Loop control Statement
1. Brack
2. Continue
3. pass


```python
for i in range(1,11):
    if i == 7:
        break
    print(i)
    
```

```python
name = ("Farhin","Rafi","Hira","Tripty")

for i in name:
    if i=="Hira":
        break
print(i)
```

```python
# continue 

for i in range(1,11):
    if i == 7:
        continue
    print(i)           
```

```python
# pass

for i in range(1,11):
    pass
```
