## Lists Methods/Functions

Lists is Mutable, That means you can change the list

```python
L.append(5)
```

```python
L = [1,2,3,4]

L[0] = 100

print(L)
```

```python
L = [1,2,3,4]

L[2] = 200

print(L)
```

```python
# editing with slicing

L = [1,2,3,4]

L[1:4] = [200,300,400]

print(L)
```

## extend

```python
L = [1,2,3,4]

a = [10,11,12]

L.extend(a)

print(L)
```

```python
L.extend("Hello")

print(L)
```

# insert

```python
L = [1,2,3,4]

L.insert(1, 200)

print(L)
```

```python
## Deleting items from a List

L = [1,2,3,4]

del L[0]

print(L)
```

```python
L = [1,2,3,4]

del L[1:3]

print(L)
```

# remove

```python
L = [1,2,3,4]

L.remove(4)

print(L)
```

# pop

```python
L = [1,2,3,4]

L.pop()
L.pop()
print(L)
```

# clear

```python
L = [1,2,3,4]

L.clear()

print(L)
```

# len/min/max/sorted

```python
L = [1,2,3,4]

print(len(L))
print(min(L))
print(max(L))
print(sorted(L,reverse=True))
```

# count

```python
L = [1,2,3,4,4,4,5,5]

L.count(4)
```

# index

```python
L = [1,2,3,4,4,4,5,5]

L.index(2)
```

# reverse

```python
L = [1,2,3,4]

L.reverse()
print(L)
```

# sort

```python
L = [4,2,1,55,10]
L.sort()
print(L)
```

# copy

```python
L = [1,2,3,4]

l1 = L.copy()

print(l1)
```

```python
l1.append(500)
```
